```python
#merge
with open("FR_category_id.json") as data_file:
    data = json.load(data_file)

etag_list = []
id_list = []
title_list = []

for i in range(31):
    try:
        etag_list.append(data['items'][i]['etag'])
        id_list.append(data['items'][i]['id'])
        title_list.append(data['items'][i]['snippet']['title'])
    except KeyError:
        etag_list.append('')
        id_list.append('')
        title_list.append('')

print('-------etag list-------')
print(etag_list)
print('-------id list-------')
print(id_list)
print('-------title list-------')
print(title_list)
```

```python
FR_category_df['etag'] = etag_list
FR_category_df['category_id'] = id_list
FR_category_df['title']= title_list

print(FR_category_df)
```
