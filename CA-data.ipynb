```python
#dependencies
import pandas as pd
from sqlalchemy import create_engine
import json
import os
```

```python
#Extract CSVs into DataFrames
csv_file = "Resources/CAvideos.csv"
CAvideos_df = pd.read_csv(csv_file)
CAvideos_df.head()
```

```python
#removed three columns as a part of the clean up activities (publish_time,thumbnail_link,description)
new_CAvideos_df= CAvideos_df[["video_id","trending_date","title","channel_title","category_id","tags","views","likes","dislikes","comment_count","comments_disabled","ratings_disabled","video_error_or_removed"]].copy()
new_CAvideos_df
```

```python
#Store JSON data into a DataFrame
json_file = "Resources/CA_category_id.json"
with open(json_file) as jsonfile:
    CA_category_json = json.load(jsonfile)
    print(json.dumps(CA_category_json,indent=2))
    CA_category_df=pd.read_json(json_file)
    CA_category_df
```

```python
with open('Resources/CA_category_id.json') as data_file:
    data = json.load(data_file)
    
    
    kind_list = []
    etag_list = []
    id_list = []
    title_list = []
    data_list = []



    for i in range(31): 
        
        try:
            kind_list.append(data['items'][i]['kind'])
            etag_list.append(data['items'][i]['etag'])
            id_list.append(data['items'][i]['id'])
            title_list.append(data['items'][i]['snippet']['title'])
        except keyerror:
            kind_list.append("")
            etag_list.append("")
            id_list.append("")
            title_list.append("")
            
print("----- kind list ------")    
print(kind_list) 
print("----- etag list -------")
print(etag_list) 
print("------ id list --------")
print(id_list) 
print("------ title list ------")
print(title_list) 
print("--------finished -------")
```

```python
CA_category_df=CA_category_df.rename(columns={"id": "category_id"})

CA_category_df
```

```python
#CA_category_df["kind"] = kind_list
CA_category_df["etag"] = etag_list
CA_category_df["category_id"] = id_list
CA_category_df["title"] = title_list

print(CA_category_df)
```

```python
# Make the id_list an integer to avoid issues when attempting merge.
CA_category_df["category_id"] = CA_category_df["category_id"].astype(int)
CA_category_df
```

```python
#merged on category id using left join
combined_CAdata=pd.merge(new_CAvideos_df,CA_category_df, on= "category_id", how = "left")
combined_CAdata
```
